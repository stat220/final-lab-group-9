```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.read_csv('https://richardson.byu.edu/220/OnlineNewsPopularity.csv')

df.columns = df.columns.str.strip()

target = 'shares'
```

EDA

```python
# Plot target variable
plt.figure(figsize=(10, 6))
sns.histplot(df[target], kde=True, bins=30, color='blue')
plt.title('Distribution of Shares', fontsize=14)
plt.xlabel('Shares', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print(df[target].describe())
```

```python
# Due to highly skewed, Log-transform the target variable to reduce skewness
df['log_shares'] = np.log1p(df['shares'])

plt.figure(figsize=(10, 6))
sns.histplot(df['log_shares'], kde=True, bins=30, color='green')
plt.title('Distribution of Log-Transformed Shares', fontsize=14)
plt.xlabel('Log(Shares)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Initial Linear Regression Model
X = df.drop(columns=[target, 'url', 'log_shares'])
y = df['shares']

# Keep numeric columns
X = X.select_dtypes(include=[np.number])

X_const = sm.add_constant(X)
ols_model = sm.OLS(y, X_const).fit()
print(ols_model.summary())
```

```python
# Higher-Order Terms
from sklearn.preprocessing import PolynomialFeatures

# Add higher-order terms (polynomial and interaction terms)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

X_poly_df = pd.DataFrame(X_poly, columns=poly.get_feature_names_out(X.columns))

X_poly_const = sm.add_constant(X_poly_df)

ols_poly_model = sm.OLS(y, X_poly_const).fit()
print(ols_poly_model.summary())
```

```python
# Regression tree for identifying important predictors
tree = DecisionTreeRegressor(random_state=123)
tree.fit(X, y)

feature_importances = pd.Series(tree.feature_importances_, index=X.columns).sort_values(ascending=False)
print("Top Predictors from the Regression Tree:")
print(feature_importances.head(10))
```

```python
# Visualize top predictors
top_features = feature_importances.index[:5]

for feature in top_features:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=df[feature], y=df[target])
    plt.title(f'{feature} vs Shares', fontsize=14)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Shares', fontsize=12)
    plt.grid()
    plt.show()
```

Linear Regression with Tuning

```python
X_final = X[top_features]
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.3, random_state=123)

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

print("Test MSE:", mean_squared_error(y_test, y_pred))
print("Test R²:", r2_score(y_test, y_pred))
```

```python
# LASSO Regression to tune ans select features
alphas = np.logspace(-3, 1, 50)
lasso = Lasso(random_state=123)
grid = GridSearchCV(lasso, {'alpha': alphas}, cv=5, scoring='neg_mean_squared_error')
grid.fit(X_train, y_train)

best_alpha = grid.best_params_['alpha']
print("Best alpha for LASSO:", best_alpha)

lasso_best = Lasso(alpha=best_alpha, random_state=123)
lasso_best.fit(X_train, y_train)

y_pred_lasso = lasso_best.predict(X_test)
print("Test MSE (LASSO):", mean_squared_error(y_test, y_pred_lasso))
print("Test R² (LASSO):", r2_score(y_test, y_pred_lasso))
```

Regression Tree Modeling
