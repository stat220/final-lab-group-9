# Mashable Lab

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.read_csv('https://richardson.byu.edu/220/OnlineNewsPopularity.csv')

df.columns = df.columns.str.strip()

target = 'shares'

df.head()
```

## EDA

### Target Variable Modification

```python
# Plot target variable
plt.figure(figsize=(10, 6))
sns.histplot(df[target], kde=True, bins=30, color='blue')
plt.title('Distribution of Shares', fontsize=14)
plt.xlabel('Shares', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print(df[target].describe())
```

<span style="color:red;">
The high skew of the shares distribution indicates that a log distribution may be more suitable.
</span>

```python
# Due to highly skewed, Log-transform the target variable to reduce skewness
df['log_shares'] = np.log1p(df['shares'])

plt.figure(figsize=(10, 6))
sns.histplot(df['log_shares'], kde=True, bins=30, color='green')
plt.title('Distribution of Log-Transformed Shares', fontsize=14)
plt.xlabel('Log(Shares)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

target = 'log_shares'
```

### Feature Significance

```python
# Initial Linear Regression Model
X_init = df.drop(columns=['log_shares', 'url', 'shares'])
y = df[target]

# Keep numeric columns
X_init = X_init.select_dtypes(include=[np.number])
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_init)
X = pd.DataFrame(X_scaled, columns=X_init.columns, index=X_init.index)

X_const = sm.add_constant(X)
ols_model = sm.OLS(y, X_const).fit()
print(ols_model.summary())
```

```python
# Higher-Order Terms
from sklearn.preprocessing import PolynomialFeatures

# Add higher-order terms (polynomial and interaction terms)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

X_poly_df = pd.DataFrame(X_poly, columns=poly.get_feature_names_out(X.columns))

X_poly_const = sm.add_constant(X_poly_df)

ols_poly_model = sm.OLS(y, X_poly_const).fit()
ols_significant_features = ols_poly_model.pvalues.sort_values().index.tolist()
print(ols_poly_model.summary())
print(ols_significant_features[:5])
```

### Feature Importance

```python
# Regression tree for identifying important predictors
tree = DecisionTreeRegressor(random_state=123)
tree.fit(X_poly_df, y)

feature_importances = pd.Series(tree.feature_importances_, index=X_poly_df.columns).sort_values(ascending=False)
print("Top Predictors from the Regression Tree:")
print(feature_importances.head(10))
```

### Combined Best Features

```python
n_features = 20
dt_important_features = list(feature_importances.index)
combined_ranks = {feature: (ols_significant_features.index(feature) + dt_important_features.index(feature) + 2) / 2 for feature in dt_important_features}

best_features = sorted(combined_ranks, key=combined_ranks.get)
top_combined_features = set(best_features[:n_features])
top_individual_features = set(ols_significant_features[:n_features//2] + dt_important_features[:n_features//2])
print(top_combined_features.symmetric_difference(top_individual_features))
print(top_combined_features.intersection(top_individual_features))
if len(top_combined_features.intersection(top_individual_features)) >= n_features / 2:
    top_features = list(top_combined_features.intersection(top_individual_features))
else:
    top_features = list(top_individual_features)
```

```python
# Visualize top predictors

for feature in top_features:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=X_poly_df[feature], y=df[target])
    plt.title(f'{feature} vs Shares', fontsize=14)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Shares', fontsize=12)
    plt.grid()
    plt.show()
    print(feature)
```

<span style="color:red;">
After visualizing the top predictors, certain of the features appear to be non-informative. We remove them.
</span>

```python
to_drop = ['n_non_stop_words', 'n_tokens_content n_non_stop_words', 'n_non_stop_words rate_positive_words']
top_features = [feature for feature in top_features if feature not in to_drop]
top_features
```

### EDA Report

## Linear Regression

### Basic Linear Regression

```python
X_final = X_poly_df[top_features]
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.3, random_state=123)

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

print("Test MSE:", mean_squared_error(y_test, y_pred))
print("Test RÂ²:", r2_score(y_test, y_pred))
```

### LASSO

```python
# LASSO Regression to tune ans select features
alphas = np.logspace(-3, 1, 50)
lasso = Lasso(random_state=123)
grid = GridSearchCV(lasso, {'alpha': alphas}, cv=5, scoring='neg_mean_squared_error')
grid.fit(X_train, y_train)

best_alpha = grid.best_params_['alpha']
print("Best alpha for LASSO:", best_alpha)

lasso_best = Lasso(alpha=best_alpha, random_state=123)
lasso_best.fit(X_train, y_train)

y_pred_lasso = lasso_best.predict(X_test)
print("Test MSE (LASSO):", mean_squared_error(y_test, y_pred_lasso))
print("Test RÂ² (LASSO):", r2_score(y_test, y_pred_lasso))
```

### Linear Regression Report

## Decision Tree Regression

### CCP Pruned Decision Tree

```python
path = tree.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas

ccp_alphas_sampled = np.linspace(ccp_alphas.min(), ccp_alphas.max(), num=10)

# Cross-validation to find the best alpha
mse_scores = []
for alpha in ccp_alphas_sampled:
    temp_tree = DecisionTreeRegressor(random_state=123, ccp_alpha=alpha)
    mse = -cross_val_score(temp_tree, X_train, y_train, cv=3, scoring='neg_mean_squared_error').mean()
    mse_scores.append(mse)

best_alpha = ccp_alphas_sampled[np.argmin(mse_scores)]

pruned_tree = DecisionTreeRegressor(random_state=123, ccp_alpha=best_alpha)
pruned_tree.fit(X_train, y_train)

y_pred_tree = pruned_tree.predict(X_test)
print("Test MSE (Pruned Tree):", mean_squared_error(y_test, y_pred_tree))
print("Test RÂ² (Pruned Tree):", r2_score(y_test, y_pred_tree))
```

### CCP Pruned Random Forest

```python
rf_model = RandomForestRegressor(n_estimators=50, random_state=123, max_depth=5, ccp_alpha=best_alpha)
rf_model.fit(X_train, y_train)

rf_predictions = rf_model.predict(X_test) #np.mean([tree.predict(X_test) for tree in pruned_trees], axis=0)

print("Test MSE (Pruned Random Forest):", mean_squared_error(y_test, rf_predictions))
print("Test RÂ² (Pruned Random Forest):", r2_score(y_test, rf_predictions))
```
